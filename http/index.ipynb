```python
# Search events for a specific tenant (need authentication and tenant in the header)
# http://localhost:8000/api/events-search/

response = requests.get(
    f"{BASE_URL}/api/events-search/",
    headers={
        "Authorization": f"Bearer  {token}",
        "X-Tenant": "dan_1766093036"
        },
    timeout=10,
    params= {
        "query": "over"
    }
)
data = response.json()

data
```

```python
# Load the specif event detail so you can start scannign its tickets
# Requires **authentication** and the **X-Tenant** variable and the **event_slug**
# http://localhost:8000/api/events/{slug}/

events = data["events"]
if not len(events) > 0:
    print("Nothing to fetch")

slug = events[0]["slug"]

response = requests.get(
    f"{BASE_URL}/api/events/{slug}/",
    headers={
        "Authorization": f"Bearer  {token}",
        "X-Tenant": "tim"
        },
    timeout=10,

)
response.json()
```

```python
# Scan a specific ticket belonging to a spaecif event
# Requires **authentication** and the **X-Tenant** variable and the **event_slug**
# and also the ticket token
# http://localhost:8000/api/events-ticket/<event_slug>/scan/

response = requests.post(
    f"{BASE_URL}/api/events-ticket/{slug}/scan/",
    headers={
        "Authorization": f"Bearer  {token}",
        "X-Tenant": "tim"
        },
    timeout=10,
    json={
        "qr_code_data": "064488388292990002"
    }
)
response.json()
```
